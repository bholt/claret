```python
def run(nshards, nclients):
    from os import system
    cmd = "ssh candy.cs 'cd tapir/src; tools/gen_config.py --nshards={s}; tools/run_ds_experiment.sh nshards={s} nclients={c} ntxns=$((8000/{c})) initposts=1000 initusers=500'".format(s=nshards, c=nclients)
    print '>>>', cmd
    system(cmd)
```

```python
valid = 'generator_time is not null and total_time is not null'
target = 3
for s in [1, 2, 4]:
    for c in [1, 2, 4, 8]:
        cond = 'nshards = %d and nclients = %d' % (s, c)
        r = [dict(x) for x in db.query('select count(*) as ct from tapir where %s and %s' % (valid, cond))]
        missing = target - r[0]['ct']
        print cond, '-> missing', missing
        for t in range(0, missing):
            run(nshards=s, nclients=c)
```
